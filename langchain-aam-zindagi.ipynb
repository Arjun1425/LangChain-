```python
import random
class Nakli_llm:

    def __init__(self):
        print("LLM is created")

    def predict(self, prompt):
        response_list = [
            "Arjun is good",
            "Car has four doors",
            "Cat is black"
        ]
        return {'response' : random.choice(response_list)}
```

```python
class Prompt:

    def __init__(self, template: str, input_variable: str):
        self.template = template
        self.input_variable = input_variable

    def format(self, input_dict):
        return self.template.format(**input_dict)
```

```python
prompt = Prompt(
    template="Write a {lenght} poem about {topic}",
    input_variable=['topic']
)

prompt.format({'lenght': 'short', 'topic': 'cricket'})
```

```python
class naklichain:

    def __init__(self, prompt, llm):
        self.prompt = prompt
        self.llm = llm

    def run(self, input_dict):
        final_prompt = self.prompt.format(input_dict)
        result = self.llm.predict(final_prompt)

        return result['response']
```

```python
llm = Nakli_llm()

template = Prompt(
    template="Write a {length} poem about {topic}",
    input_variable=['topic']
)

chain = naklichain(template, llm)
chain.run({'length': 'short', 'topic': 'football'})
```

### Here comes Runnable class

```python
from abc import ABC, abstractmethod

class Runnable(ABC):
    
    @abstractmethod
    def invoke(self, input_dict):
        pass
```

```python
class Nakli_llm(Runnable):

    def __init__(self):
        print("LLM is created")

    def invoke(self, prompt):
        response_list = [
            "Arjun is good",
            "Car has four doors",
            "Cat is black"
        ]
        return {'response' : random.choice(response_list)}

    def predict(self, prompt):
        response_list = [
            "Arjun is good",
            "Car has four doors",
            "Cat is black"
        ]
        print("This method is going to be deprecated, use invoke() instead.")
        return {'response' : random.choice(response_list)}

class Nakli_Prompt(Runnable):

    def __init__(self, template: str, input_variable: str):
        self.template = template
        self.input_variable = input_variable

    def invoke(self, input_dict):
        return self.template.format(**input_dict)

    def format(self, input_dict):
        print("This method is going to be deprecated, use invoke() instead.")
        return self.template.format(**input_dict)

class RunnableConnector(Runnable):

    def __init__(self, runnable_list):
        self.runnable_list = runnable_list

    def invoke(self, input_data):
        for runnable in self.runnable_list:
            input_data = runnable.invoke(input_data)

        return input_data
    
class NakliStrOutParser(Runnable):

    def __init__(self):
        pass

    def invoke(self, input_data):
        return input_data['response']
```

```python
llm = Nakli_llm()

template = Nakli_Prompt(
    template="Write a {length} poem about {topic}",
    input_variable=['length', 'topic']
)

parser = NakliStrOutParser()

chain = RunnableConnector([template, llm, parser])

chain.invoke({'length': 'short', 'topic': 'india'})
```

```python
template_1 = Nakli_Prompt(
    template = "Write me a joke on {subject}",
    input_variable = ['subject']
)

template_2 = Nakli_Prompt(
    template = "Explain me this joke: {response}",
    input_variable = ['response']
)

llm = Nakli_llm()
parser = NakliStrOutParser()

chain_1 = RunnableConnector([template_1, llm])

chain_2 = RunnableConnector([template_2, llm, parser])

#chain_2.invoke({'joke': 'Why did the chicken cross the road? To get to the other side!'})

final_chain = RunnableConnector([chain_1, chain_2])

final_chain.invoke({'subject': 'chicken'})
```
