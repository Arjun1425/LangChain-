## Build-in Tools in LangChain

### DuckDuckGo

```python
from langchain_community.tools import DuckDuckGoSearchResults

search_tool = DuckDuckGoSearchResults()

query = "Recent news of USA"

result = search_tool.invoke(query)

print(result)
```

### ShellTool

```python
from langchain_community.tools import ShellTool

shell_tool = ShellTool()

cmd = "pwd"
result = shell_tool.invoke(cmd)
print(result)
```

## Custom tools

* Create a function

```python
from langchain_community.tools import tool
```

```python
def add_numbers(a, b):
    "Add two numbers and return the result."
    return a + b
```

* add type hints

```python
def add_numbers(a: int, b: int) -> int:
    "Add two numbers and return the result."
    return a + b
```

* add tool decorator

```python
@tool
def add_numbers(a: int, b: int) -> int:
    "Add two numbers and return the result."
    return a + b
```

```python
result = add_numbers.invoke({"a": 5, "b": 10})
print(result)
```

```python
print(add_numbers.name)
print(add_numbers.description)
print(add_numbers.args)
```

* What does LLM see, when you send this tool to LLM? --> Json_schema

```python
print(add_numbers.args_schema.model_json_schema())
```
