```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

load_dotenv()

model = ChatOpenAI(model = "gpt-4o")

result = model.invoke("What is the capital of United States?")

print(result)
```

```python
print(result.content)
```

# Temperature

**set 0 to 0.3 for factual answers (Math, code)**

**set 0.5 to 0.7 for balanced answers (general QA)**

**set 0.9 to 1.2 for creative writting**

**set 1.5+ Maximum randomness** 

```python
result1 = model.invoke("Tell me a one line joke on people doing PhD in computational chemistry", temperature = 0.1)
result2 = model.invoke("Tell me a one line joke on people doing PhD in computational chemistry", temperature = 1.8)

result1_ = model.invoke("Tell me a one line joke on people doing PhD in computational chemistry", temperature = 0.1)
result2_ = model.invoke("Tell me a one line joke on people doing PhD in computational chemistry", temperature = 1.8)
```

```python
print(f"result1: {result1.content}")
print(f"result2: {result2.content}")
print(f"result1_: {result1_.content}")
print(f"result2_: {result2_.content}")
```
